```python
from dipy.io.streamline import load_tractogram
import numpy as np
import nibabel as nib
from dipy.io.image import load_nifti
from dipy.io.utils import get_reference_info
from dipy.tracking import utils
import matplotlib.pyplot as plt
```

Load tractogram and atlas

```python
tractogram = load_tractogram('Track0.trk', 'same')
```

```python
tracks_affine, dims, voxel_szs, voxel_ords = get_reference_info(tractogram)
```

```python
print(tracks_affine)
```

```python
atlas_img, atlas_affine = load_nifti('inverted_atlas.nii.gz')
```

```python
print(atlas_affine)
```

```python
atlas_img.shape
```

Good sign is that both affines and dimensions are the same.

```python
M, grouping = utils.connectivity_matrix(tractogram.streamlines,tracks_affine, label_volume=atlas_img.astype(np.uint8), return_mapping=True, mapping_as_streamlines=True)
```

```python
plt.imshow(np.log1p(M), interpolation='nearest',)
```
